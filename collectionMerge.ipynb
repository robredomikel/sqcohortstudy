```python
"""
PARAMS: nonSQProjects - List of of non-SQ project names
        SQProjects - List of of SQ project names
        asfProjectNames - Initial dataframe with the up-to-date analysis variables
RETURNS: A combined dataframe with information gathered for the number of commits and developers before the follow-up period in each project
         added to the Boolean informing whether the project uses SQ or not, all added to the initially existing dataframe.
"""
def beforeFollowUp(nonSQProjects, SQProjects, asfProjectNames):

    colnames = ['full_name', '#commitsBeforeFollowUp', "#DevsBeforeFollowUp", 'SQ/NonSQ']
    df = pd.DataFrame(columns=colnames)

    nonSQPath = "/home/mikel/Desktop/project1/testCommits/commitsBeforeFollowUp/nonSQprojects/"
    SQPath = "/home/mikel/Desktop/project1/testCommits/commitsBeforeFollowUp/SQprojects/"

    emptyProjects = []

    # Addition of non-SQ projects' confounders into a subcommon dataframe
    for nonSQ in nonSQProjects:

        print(nonSQ)
        temp_df = pd.read_csv(nonSQPath + nonSQ)
        # Add at the end of the df the unique number of developers in the project before the follow-up period.
        temp_df.at[len(temp_df.index) - 1, 'authorID'] = len(temp_df['authorID'].unique())
        try:
            # Get the values from the end row of the dataframe.
            df.loc[len(df.index)] = [temp_df.iloc[-2:-1]['project_fullName'].values[0], temp_df.iloc[-1:]['commit_sha'].values[0], temp_df.iloc[-1:]['authorID'].values[0], 'NonSQ']
        except: # Projects that don't have any commit registered in Github before the follow-up period (Hence, they neither have developers)
            df.loc[len(df.index)] = ["apache/" + nonSQ[:-4], 0, 0, 'NonSQ']
            emptyProjects.append(nonSQ)

    # Addition of SQ projects' confounders into a subcommon dataframe
    for SQ in SQProjects:

        print(SQ)
        temp_df = pd.read_csv(SQPath + SQ)
        # Add at the end of the df the unique number of developers in the project before the follow-up period.
        temp_df.at[len(temp_df.index) - 1, 'authorID'] = len(temp_df['authorID'].unique())
        try:
            # Get the values from the end row of the dataframe.
            df.loc[len(df.index)] = [temp_df.iloc[-2:-1]['project_fullName'].values[0], temp_df.iloc[-1:]['commit_sha'].values[0], temp_df.iloc[-1:]['authorID'].values[0], 'SQ']
        except: # Projects that don't have any commit registered in Github before the follow-up period ((Hence, they neither have developers))
            df.loc[len(df.index)] = ["apache/" + SQ[:-4], 0, 0, 'SQ']
            emptyProjects.append(SQ)

    df['#DevsBeforeFollowUp'] = df['#DevsBeforeFollowUp'].convert_dtypes()

    asfCombined2 = pd.merge(asfProjectNames, df, on='full_name', how='inner')
    print(f"Length of the dataset after merge of new variables: {len(asfCombined2)}")
    asfCombined2 # 2373 (2 repos are lost in the merge)

    return asfCombined2
```

# Check the age of the projects at the beginning of the follow-up (DONE)

```python
"""
PARAMS: nonSQProjects - List of of non-SQ project names
        SQProjects - List of of SQ project names
        asfCombined3 - Previous dataframe with the up-to-date analysis variables
RETURNS: A combined dataframe with information gathered for the age of each project at the begining of the follow-up period added to the initially existing dataframe.
"""
def ageCalculation(nonSQProjects, SQProjects, asfCombined3):

    colnames = ['full_name', 'age_in_ASF']
    df = pd.DataFrame(columns=colnames)

    nonSQPath = "/home/mikel/Desktop/project1/testCommits/commitsBeforeFollowUp/nonSQprojects/"
    SQPath = "/home/mikel/Desktop/project1/testCommits/commitsBeforeFollowUp/SQprojects/"

    emptyProjects = []

    # Addition of non-SQ projects' confounders into a subcommon dataframe
    for nonSQ in nonSQProjects:

        print(nonSQ)
        temp_df = pd.read_csv(nonSQPath + nonSQ)
        try:
            last_date = pd.Timestamp(temp_df['author_date'].iloc[pd.to_datetime(temp_df['author_date'], utc=True).argmax()])
            first_date = pd.Timestamp(temp_df['author_date'].iloc[pd.to_datetime(temp_df['author_date'], utc=True).argmin()])
            df.loc[len(df.index)] = [temp_df.iloc[-2:-1]['project_fullName'].values[0], last_date - first_date]
        except: # Projects that don't have any commit registered in Github before the follow-up period
            df.loc[len(df.index)] = ["apache/" + nonSQ[:-4], 0]

    # Addition of SQ projects' confounders into a subcommon dataframe
    for SQ in SQProjects:

        print(SQ)
        temp_df = pd.read_csv(SQPath + SQ)

        try:
            last_date = pd.Timestamp(temp_df['author_date'].iloc[pd.to_datetime(temp_df['author_date'], utc=True).argmax()])
            first_date = pd.Timestamp(temp_df['author_date'].iloc[pd.to_datetime(temp_df['author_date'], utc=True).argmin()])
            df.loc[len(df.index)] = [temp_df.iloc[-2:-1]['project_fullName'].values[0], last_date - first_date]
        except: # Projects that don't have any commit registered in Github before the follow-up period
            df.loc[len(df.index)] = ["apache/" + SQ[:-4], 0]

    asfCombined4 = pd.merge(asfCombined3, df, on='full_name', how='inner')
    print(f"Length of the dataset after the merge process: {len(asfCombined4)}") # 2372

    return asfCombined4
```

# Pipeline to merge the issue data from Jira and Github into same files (DONE):

```python
"""
PARAMS: githubIssues - dataframe with registered issues from projects in GitHub
        jiraIssues -  dataframe with registered issues from projects in Jira
RETURNS: clean_listJira - List of projects in Jira identified as Mature
         clean_listGitHub - List of projects in GitHub identified as Mature
         incubator_retired - List of pruned projects that are part of the incubator retired group in ASF
         incubator - List of pruned projects that form part of the incubator group in ASF
         attic - List of pruned projects that form part of the Attic group in ASF
         null_projects - List of unidentified projects from Jira (Do not exist in GitHub)
         incubator_GitHub - List of pruned projects from GitHub that form part of the incubator group in ASF
         incubator_Jira - List of pruned projects from Jira that form part of the incubator group in ASF
"""
def nonMaturePruning(githubIssues, jiraIssues):

    attic = []
    incubator_Jira = []
    incubator_GitHub = []
    incubator_retired_Jira = []
    incubator_retired_GitHub = []
    null_projects = []
    clean_listJira = [] # Projects that can be merged from Jira
    clean_listGitHub = [] # Projects that can be merged from GitHub

    for project in jiraIssues:

        splitList = project.split('-')
        # If the project is in the attic:
        if splitList[-1] == 'attic.csv':
            attic.append(project)
            print(f"{project} goes into Attic")
        elif len(splitList) >= 2 and splitList[0] == 'incubator':
            incubator_Jira.append(project)
        # If the project is in the incubator retired group
        elif len(splitList) >= 3 and splitList[1] == 'retired':
            incubator_retired_Jira.append(project)
            print(f"{project} goes into incubator retired")
        # If the project hasn't been identified in GitHub nor in the ASF webpage
        elif splitList[-1] == 'null.csv':
            null_projects.append(project)
            print(f"{project} hasn't been identified")
        else:
            clean_listJira.append(project)
            # print(f"{project} is clean. Ready to process!")

    for project in githubIssues:

        splitList = project.split('-')

        # If the project is in the incubator.
        if len(splitList) >= 2 and splitList[0] == 'incubator':
            incubator_GitHub.append(project)
        # If the project is in the incubator retired group
        elif len(splitList) >= 3 and splitList[1] == 'retired':
            incubator_retired_GitHub.append(project)
        else:
            clean_listGitHub.append(project)

    incubator = [] # Incubator with unique repos from Jira and GitHub
    incubator_retired = [] # Incubator retired repos from Jira and GitHub.

    # Unique values in incubator_retired
    for i in incubator_retired_GitHub: i if i in incubator_retired else incubator_retired.append(i)
    for i in incubator_retired_Jira: i if i in incubator_retired else incubator_retired.append(i)

    # Unique values in incubator
    for i in incubator_GitHub: i if i in incubator else incubator.append(i)
    for i in incubator_Jira: i if i in incubator else incubator.append(i)

    return clean_listJira, clean_listGitHub, incubator_retired, incubator, attic, null_projects, incubator_GitHub, incubator_Jira
```

```python
"""
PARAMS: pruningResults - List of projects resulting from the nonMaturePruning() function
        commonFolderPath - Path of folder where all the suitable projects and their issues will be stored

RETURNS: matching_projects - Same projects with issues registered in Jira and GitHub.
         no_matching - No matching projects with issues in GitHub (meaning that they have commit data so that they are not pruned)
         empty_repos - Repositories with no issues after the time based pruning of issues
        nomatching_Jira - No matching projects from Jira that stand without commit information since they are not linked way any project in GitHub.
"""
def combineIssues(pruningResults,commonFolderPath):

    # SECOND: Merge projects from GitHub and Jira and create a file with the combined issues.

    # There may be projects that have the same filename, fetch them and create a new csv file with the necessary variables
    matching_projects = []
    no_matching = []

    # EMPTY DATAFRAMES WILL BE ALSO DISCARDED AND COUNTED.
    empty_repos = []

    # For the sake of readability:
    clean_listJira = pruningResults[0]
    clean_listGitHub = pruningResults[1]

    for file in clean_listGitHub:

        df_headers = ['full_name', 'state', 'created_at', 'closed_at']
        jointDF = pd.DataFrame(columns=df_headers)

        # Check if it exists in Jira, in order to merge the same repo issues from Jira and Github:
        if file in clean_listJira:
            print(file)
            jiradf = jointDF.copy()
            # First, deal with Jira data:
            df = pd.read_csv(f'/home/mikel/Desktop/project1/JiraDataCrawler/1.DataDownload/jiraProjectIssues/{file}')

            if df.empty:
                empty_repos.append(file)
                continue
            try:
                jiradf['created_at'] = pd.to_datetime(df['created_at'], errors='coerce').dt.tz_convert(None)
                jiradf['closed_at'] = pd.to_datetime(df['resolution_at'], errors='coerce').dt.tz_convert(None)
            except TypeError:
                empty_repos.append(file)
                continue
            jiradf['state'] = df['state'].replace(to_replace='resolved', value='closed')
            jiradf['full_name'] = [f'apache/{file[:-4]}' for i in range(len(df))]

            # Add also the data from gitHub:
            gitHubdf = jointDF.copy()
            dfGH = pd.read_csv(f'/home/mikel/Desktop/project1/githubIssueTables(TimeStamp)/{file}')
            gitHubdf['created_at'] = dfGH['created_at']
            gitHubdf['closed_at'] = dfGH['closed_at']
            gitHubdf['state'] = dfGH['state']
            gitHubdf['full_name'] = [f'apache/{file[:-4]}' for i in range(len(dfGH))]

            result = pd.concat([jiradf, gitHubdf], join='outer', ignore_index='True')

            result.to_csv(f'{commonFolderPath}/{file}')

            matching_projects.append(file)

        # If the file in GitHub isn't in Jira. Only process that one.
        else:

            gitHubdf = jointDF.copy()
            dfGH = pd.read_csv(f'/home/mikel/Desktop/project1/githubIssueTables(TimeStamp)/{file}')

            if dfGH.empty:
                empty_repos.append(file)
                continue

            gitHubdf['created_at'] = dfGH['created_at']
            gitHubdf['closed_at'] = dfGH['closed_at']
            gitHubdf['state'] = dfGH['state']
            gitHubdf['full_name'] = [f'apache/{file[:-4]}' for i in range(len(dfGH))]

            gitHubdf.to_csv(f"{commonFolderPath}/{file}")

            no_matching.append(file)

    nomatching_Jira = []
    for project in clean_listJira:
        if project not in matching_projects:
            nomatching_Jira.append(project)

    return matching_projects, no_matching, empty_repos, nomatching_Jira
```

```python
"""
PARAMS: githubIssues - dataframe with registered issues from projects in GitHub
        jiraIssues - dataframe with registered issues from projects in Jira
        commonFolderPath - Path of folder where all the suitable projects and their issues will be stored

Performes the previous two functions' activity reporting the outcome of the entire process.
"""
def mergeProcess(githubIssues, jiraIssues, commonFolderPath):

    # FIRST PRUNING STEP:
    pruningResults = nonMaturePruning(githubIssues, jiraIssues)
    # SECOND PRUNING STEP + MERGE OF FILES
    pruningResults2 = combineIssues(pruningResults, commonFolderPath)

    print('DESCRIPTIVE NUMBERS OF THE PROCESS:')

    print(f'Initial number of GitHub repos: {len(githubIssues)}')
    print(f'Initial number of Jira repos: {len(jiraIssues)}')
    print('')
    print(f"Amount of projects in Attic from Jira: {len(pruningResults[4])}")
    print(f"Amount of projects from the incubator: {len(pruningResults[3])}")
    print(f"-----Amount of projects from the incubator in GitHub: {len(pruningResults[6])}")
    print(f"-----Amount of projects from the incubator in Jira: {len(pruningResults[7])}")
    #print(f"Amount of projects from the incubator that got retired: {len(incubator_retired)}")
    print(f"Amount of projects non classified projects: {len(pruningResults[5])}")
    print(f"Amount of clean projects from Jira: {len(pruningResults[0])}")
    print(f"Amount of clean projects from GitHub: {len(pruningResults[1])}")
    print('')
    print(f"AMOUNT OF PROJECTS MATCHING JIRA & GITHUB: {len(pruningResults2[0])}")
    print(f"AMOUNT OF PROJECTS NOT MATCHING FROM GITHUB: {len(pruningResults2[1])}")
    print(f"AMOUNT OF PROJECTS NOT MATCHING FROM JIRA: {len(pruningResults2[3])}")

    commonIssues = os.listdir(commonFolderPath)
    print(f"PROJECT REPOS THAT WERE EMPTY DURING THE PROCESSING: {len(pruningResults2[2])}")
    print(f"TOTAL PROJECTS TO CALCULATE VELOCITY: {len(commonIssues)} (empty (from issues) ones have been excluded)")
```

```python
# FUNCTION CALL:
commonFolderPath = "/home/mikel/Desktop/project1/commonIssueFolder"
githubIssues = os.listdir('/home/mikel/Desktop/project1/githubIssueTables(TimeStamp)') # Length = 2375
jiraIssues = os.listdir('/home/mikel/Desktop/project1/JiraDataCrawler/1.DataDownload/jiraProjectIssues') # Length = 483

mergeProcess(githubIssues, jiraIssues, commonFolderPath)
```
